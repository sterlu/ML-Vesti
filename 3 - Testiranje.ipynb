```python
import numpy as np

def generate(model, tokenizer):
    print()
    print('###### Standard decoding ######')
    for temp in np.linspace(0.7, 1, 4):
        print()
        print("###### Temperature: " + str(temp))
        for i in range(0, 5):
            out = model.generate(
                do_sample=True,
                temperature=temp,
                max_length=50,
                min_length=5,
            )[0]
            print(tokenizer.decode(out, skip_special_tokens=True))

def generate_beam(model, tokenizer):
    print()
    print('###### Beam search ######')
    for temp in np.linspace(0.7, 1.2, 6):
        print()
        print("###### Temperature: " + str(temp))
        for i in range(0, 5):
            out = model.generate(
                # tokenizer.encode('Samo još danas', return_tensors='pt'),
                do_sample=True,
                temperature=temp,
                max_length=30,
                num_beams=3,
                no_repeat_ngram_size=2,
                num_return_sequences=3,
                early_stopping=True
                # min_length=5,
            )[0]
            print(tokenizer.decode(out, skip_special_tokens=True))

def generate_top_p(model, tokenizer):
    print()
    print('###### Top-P + Top-K ######')
    for temp in np.linspace(0.7, 1, 4):
        print()
        print("###### Temperature: " + str(temp))
        for i in range(0, 5):
            out = model.generate(
                do_sample=True,
                temperature=temp,
                max_length=50,
                min_length=5,
                # top_k=50,
                top_p=0.90,

            )[0]
            print(tokenizer.decode(out, skip_special_tokens=True))
```

```python
import sys
in_colab = 'google.colab' in sys.modules
model_dir = '/content/drive/My Drive/modeli/' if in_colab else './modeli/'

from transformers import AutoModelForCausalLM, AutoModelForMaskedLM

def sample_all(model_name):
    from transformers import PreTrainedTokenizerFast
    tokenizer = PreTrainedTokenizerFast(tokenizer_file=model_dir + model_name + "/tokenizer.json")
    tokenizer.pad_token = "<pad>"
    model = AutoModelForCausalLM.from_pretrained(model_dir + model_name)
    model.config.pad_token_id = 1
    generate(model, tokenizer)
    generate_beam(model, tokenizer)
    generate_top_p(model, tokenizer)
```

```python
sample_all('gpt2_4')
```
